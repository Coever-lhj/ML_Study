###  linearly separable ?? Desicion Boundary??



$ \ $ 선은 데이터 집합을 분리하는 데 사용되며, 이를 **초평면 분리(Separating hyperplane)**라고 함.



+ 2차원에서는 하나의 선이 된다. 

+ 3차원 데이터 집합을 가지고 있다면, 이 데이터를 분리하기 위해서는 하나의 평면이 필요하게 된다. ......

+ 일반화 시켜보면 N차원의 데이터라면 **"N-1"** 이라는 **초평면**이라고 불리며, **의사 결절 범위(Desicion Boundary)**가 된다.


$ \ $ 의사 결정 범위 로부터 멀리 떨어져 있는 데이터 점과 우리가 확인하고자 하는 예측에 대해 더욱더 확신할 수 있도록 분류기를 만들고자 한다. [그림 6.2]의 프레임 B~D의 플롯을 생각해 보자. 이들은 모든 데이터를 분리한다. 그런데, 이 중 가장 좋은 것은 어떤 것일까? 

+ *초평면 분리에 대한 데이터의 평균 거리를 최소화해야 할까??* NO!! 

+ 초평면 분리에 가장 가까운 지점을 찾기를 원하면서, 이 지점이 가능한 한 분리선에서 멀리 떨어져 있기를 원한다.* 이것은 **MARGIN** 이라고 함.

+ 즉, 마진이 가능한 가장 크기를 원한다. 왜냐하면, 우리가 실수를 하거나 한정된 데이터로 분류기를 훈련시키고자 할 때에도 분류기가 최대한 잘 작동하도록 튼튼하기를 원하기 때문이다.

> 초평면 분리에 가장 가까운 지점이 **SUPPORT VECTOR**이다. 지금까지 분리선에서 지지벡터까지의 거리를 가장 크게 해야 한다는 것을 알았으니

## 6.2 최대 마진 찾기

이 선이 데이터를 가장 잘 분리하는 선인지 어떻게 측정할 수 있을까? [그림 6.3]을 보면, 초평면 분리는 $W^Tx + b$ 로부터 얻는다.  

$ \ $ 점 A에서 분리면까지의 거리를 구하고자 한다면, 선에 직각 또는 수직인 선의 길이를 측정해야 한다. 이러한 측정은 다음과 같이 구할 수 있다.

$$
\frac{|w^Tx + b|}{||w||}
$$

$ \ $ 여기서 상수 $b$는 단지 로지스틱 회귀에서 $w_0$처럼 오프셋(offset)이다. 여기서 $w$와 $b$는 데이터를 위한 분리선 또는 초평면을 묘사하는 것이다.

<img src = "margin.jpg" width="300" height="300">

Decision Boundry? 를 서포트하는 벡터들을 찾아내는게 이것의 알고리즘!!

### 6.2.1 분류기 관점에서의 최적화 문제 구성하기




#### 질문!! 로지스틱 회귀에서는 분류 항목이 0 또는 1이지만 , SVM은 왜 $y_j = \pm 1$인 것일까?

#### 답변!: 단지 부호만 다르기 때문에, $(wx_j + b)y_j $ > 0 라는 방정식으로 표현할 수 있다.

### Margin

$$
d_+ = \frac{|w^Tx_+ + b|}{||w||} = \frac{+1}{||w||}
$$

$$
d_- = \frac{|w^Tx_- + b|}{||w||} = \frac{-1}{||w||}
$$

우리가 필요한 것은 "거리"라는 절댓값이므로 Margin은 위 둘의 절대값을 더함.

### $$
Margin = |d_+| + |d_-| = \frac{2}{||w||}
$$

### 목적식과 제약식 정의

SVM의 목적은 마진을 최대화하는 경계면을 찾는 것입니다. 계산상 편의를 위해 마진 절반을 제곱한 것에 역수를 취한 뒤 그 절반을 최소화하는 문제로 바꾸겠습니다. 이렇게 해도 문제의 본질은 바뀌지 않습니다.

### $$
Max \frac{2}{||w||} \rightarrow Min \frac{||w||}{2}^2
$$


여기엔 다음과 같은 제약조건이 관측치 개수만큼 붙습니다. 식의 의미는 이렇습니다. plus-plane보다 위에 있는 관측치들은 y=1이고 $w^Tx+b$가 1보다 큽니다. 반대로 minus-plane보다 아래에 있는 점들은 y=−1이고 $w^Tx+b$가 -1보다 작습니다. 이 두 조건을 한꺼번에 묶으면 아래와 같은 제약식이 됩니다.

### $$
y_i(w^Tx_i+b) ≥ 1
$$




### 라그랑지안 문제로 변환

라그랑지안 승수법(Lagrange multiplier method)은 제약식에 형식적인 라그랑지안 승수를 곱한 항을 최적화하려는 목적식에 더하여, 제약된 문제를 제약이 없는 문제로 바꾸는 기법


#### 이미 정의한 목적식과 제약식을 라그랑지안 문제로 식을 다시 쓰면 다음과 같습니다.

####  Object Function (Primal)

### $$
Min \ L_p(w,b,α_i) = \frac{1}{2}||w||^2 − \sum_{i=1}^n α_i(y_i(w^Tx_i+b)−1)
$$

#### S.T

### $$
α_i≥0, \ i=1,...,n
$$

### Dual 문제로 변환

KKT 조건에서는 Lp를 미지수로 각각 편미분한 식이 0이 되는 지점에서 최소값을 갖습니다. 다음과 같습니다.


$$
\frac{d \ L_p(w, b, \alpha_i)}{d \ w} = 0  \ \rightarrow \ w = \sum_{i=1}^n \alpha_i y_i x_i
$$

$$
\frac{d \ L_p(w, b, \alpha_i)}{d \ b} = 0  \ \rightarrow \ \sum_{i=1}^n \alpha_i y_i = 0
$$

위 식을 L에 넣어 정리해 보겠습니다.

### SVM의 해

$ \ $ 우리가 찾고자 한 답은 마진이 최대화된 분류경계면 $w^Tx+b$입니다. w와 b를 찾으면 SVM의 해를 구할 수 있게 됩니다. KKT 조건을 탐색하는 과정에서 w는 다음과 같이 도출됐습니다.

$$
w = \sum_{i=1}^n \alpha_i y_i x_i
$$

$ \ $ $x_i$와 $y_i$는 우리가 가지고 있는 학습데이터이므로 라그랑지안 승수인 α값들만 알면 w를 찾을 수 있습니다. 그런데 여기에서 $α_i$가 0인 관측치들은 분류경계면 형성에 아무런 영향을 끼치지 못합니다. 바꿔 말해 i번째 관측치에 대응하는 라그랑지안 승수 $α_i$ 가 0보다 커야 마진 결정에 유의미하다는 이야기입니다.


$ \ $ $\alpha_i$은 위 식을 QP로 풀면된다. 그러나 QP가 SVM 계산중 가장 무거운 부분이다. 그래서 QP를 최적화하려느 연구도 있었다. 그중 하나가 SMO 이다.

# 6.3 SMO 알고리즘으로 효율적인 optimization하기

## 6.3.1 John Platt 의 SMO 알고리즘

#### SVM 알고리즘 중 가장 널리 쓰이는 SMO(Sequential minimal optimization) 알고리즘

> **방법**
>
> 큰 문제를 작은 문제들로 나누며, 이렇게 나뉜 문제들을 쉽고 순차적으로 해결하여 큰 문제를 한번에 해결하는 것과 같은 답을 얻음
>
> **장점**
>
> 처리 시간 감소

> **원리**
>
> SMO 알고리즘은 $\alpha_i$와 $b$의 집합을 찾는 것이다. 알파의 집합을 구하게 되면 가중치 $w$는 쉽게 계산되며 분리 초평면을 구할 수 있다.
>
> 각각의 사이클을 최적화하기 위해서 두 개의 $\alpha$를 선택하나. 적당한 $\alpha$의 쌍을 찾게 되면, 그 중 하나는 값을 증가시키고 다른 하나는 줄인다. 적당한 $\alpha$를 구하기 위해서는 $\alpha$집합이 조건을 정확하게 충족해야 한다. 그 조건 중 하나가 $\alpha$ 쌍 모두가 그들의 마진 경계 밖에 있어야 한다는 것이다. 두 번째 조건은 $\alpha$가 이미 고정되어 있거나 경계를 갖지 않아야 한다는 것이다.

## 6.3.2 간략한 형태의 SMO로 적은 양읟 데이터 집합 해결하기

$ \ $ 완전한 SMO 알고리즘은 많은 코드를 가지고 수행된다. **완전한 SMO 알고리즘**의 원리를 이해하기 위해 간소화시킬 시킴. 

$ \ $ **간략한 형태의 SMO 알고리즘** 작업을 얻고나면 **완전한 SMO 알고리즘**을 확인해 볼 것이다. **간략한 형태의 SMO 알고리즘**은 적은 양의 코드를 사용하지만 실행 시간이 오래 걸린다. SMO 알고리즘의 바깥 반복문에서는 최적화를 위해 가장 좋은 $\alpha$를 결정한다. 간략한 형태의 알고리즘에서는 이 부분을 건너뛰고 데이터 집합에 있는 모든 $\alpha$를 대상으로 알파의 쌍을 먼저 선택하게 된다. 그런 다음, 남아 있는 $\alpha$들에서 임의로 두 번째 알파를 선택한다. 여기서 주의해야 할 점은 두 $\alpha$를 동시에 변경해야 한다는 것이다. 

> 그 이유는 다음과 같은 제약 조건이 있기 때문이다.

$$
\sum_1^n \alpha_i \cdot y^{(i)} = 0
$$

> $\alpha$ 하나만을 변경하는 것은 제약 조건을 위반하는 것으로 항상 $\alpha$ 두 개를 한꺼번에 변경한다.

$ \ $ 이를 수행하기 위해 도움 함수를 생성할 것이며, 이 함수는 범위 내에서 임의로 하나의 정수를 선택한다. 또한 너무 큰 값을 얻지 않도록 값을 고정하는 도움 함수도 필요하다.

#### [loadDataSet] txt 파일을 읽어 데이터 및 분류값 을 행렬로 구문 분석해주는 함수

```python
from numpy import *
from time import sleep

# 데이터 읽어오는 함수
def loadDataSet(fileName):
    dataMat = []; labelMat = []
    fr = open(fileName)
    for line in fr.readlines():
        lineArr = line.strip().split('\t')
        dataMat.append([float(lineArr[0]), float(lineArr[1])])
        labelMat.append(float(lineArr[2]))
    return dataMat,labelMat
```

#### [selectJrand] 

```python
# i, m을 처리하는 함수
# i => 첫번째 alpha에 대한 index
# m => alpha 전체 갯수
def selectJrand(i,m):
    j=i # we want to select any J not equal to i
    while (j==i):
        j = int(random.uniform(0,m))
    return j
```

### [Listing 6.2] 간략한 형태의 SMO 알고리즘

```python
def smoSimple(dataMatIn, classLabels, C, toler, maxIter):
    
    dataMatrix = mat(dataMatIn)
    labelMat = mat(classLabels).transpose()
    
    b = 0
    m,n = shape(dataMatrix) # (100, 2) -> m = 100, n = 2
    alphas = mat(zeros((m,1))) # alpha를 위한 column을 생성하고 0 으로 초기화 
    iter = 0 # 어떠한 alpha의 변경도 없이 데이터 집합을 처리하는 반복 횟수를 세는 변수!
    while (iter < maxIter): # maxIter 에 도달하면 중단!
        alphaPairsChanged = 0 # 어떤 alpha의 최적화를 시도할 경우 이를 기록하는 데 사용!
                              # 반복문 마지막에 확인 가능!
        
        for i in range(m):
            # 분류항목에 대한 우리의 예측인 fXi를 계산!
            fXi = float(multiply(alphas,labelMat).T*(dataMatrix*dataMatrix[i,:].T)) + b
            # 그런 다음, 오류 Ei는 예측과 이 사례의 실제 분류 항목을 기반으로 계산하게 된다.
            Ei = fXi - float(labelMat[i])#if checks if an example violates KKT conditions
            
            # 이렇게 계산된 오류가 크다면, 이 데이터 사례에 해당하는 alpha는 최적화될 수 있다.
            
            # 알파가 변경될 수 있다면 최적화 입력 (위의 조건을 여기서 검토)
            # if 문에서 + Margin과 - Margin을 둘다 검토 !
            if ((labelMat[i]*Ei < -toler) and (alphas[i] < C)) or ((labelMat[i]*Ei > toler) and (alphas[i] > 0)): 
            # 여기서 alpha가 0 또는 C 와 같지 않음을 확인! -> 0 또는 C일때 최적화할 필요 X    
                # 임의로 두 번째 alpha 선택! 위와 같이 오류 계산
                j = selectJrand(i,m)
                fXj = float(multiply(alphas,labelMat).T*(dataMatrix*dataMatrix[j,:].T)) + b
                Ej = fXj - float(labelMat[j])
                
                # alphap[i]와 alphap[j]의 복사본 생성함 
                # 이유: 수에 새로운 alpha와 이전 alpha를 비교하기 위함
                alphaIold = alphas[i].copy()
                alphaJold = alphas[j].copy();
                
                # alpha가 0과 C(입력상수) 사이의 값이 되도록 함!
                if (labelMat[i] != labelMat[j]):
                    L = max(0, alphas[j] - alphas[i])
                    H = min(C, C + alphas[j] - alphas[i])
                else:
                    L = max(0, alphas[j] + alphas[i] - C)
                    H = min(C, alphas[j] + alphas[i])
                if L==H: print("L==H"); continue
                eta = 2.0 * dataMatrix[i,:]*dataMatrix[j,:].T - dataMatrix[i,:]*dataMatrix[i,:].T - dataMatrix[j,:]*dataMatrix[j,:].T
                
                if eta >= 0: print("eta>=0"); continue
                alphas[j] -= labelMat[j]*(Ei - Ej)/eta
                alphas[j] = clipAlpha(alphas[j],H,L)
                
                if (abs(alphas[j] - alphaJold) < 0.00001): print("j not moving enough"); continue
                alphas[i] += labelMat[j]*labelMat[i]*(alphaJold - alphas[j])#update i by the same amount as j
                                                                        #the update is in the oppostie direction
                b1 = b - Ei- labelMat[i]*(alphas[i]-alphaIold)*dataMatrix[i,:]*dataMatrix[i,:].T - labelMat[j]*(alphas[j]-alphaJold)*dataMatrix[i,:]*dataMatrix[j,:].T
                b2 = b - Ej- labelMat[i]*(alphas[i]-alphaIold)*dataMatrix[i,:]*dataMatrix[j,:].T - labelMat[j]*(alphas[j]-alphaJold)*dataMatrix[j,:]*dataMatrix[j,:].T
                
                if (0 < alphas[i]) and (C > alphas[i]): b = b1
                elif (0 < alphas[j]) and (C > alphas[j]): b = b2
                else: b = (b1 + b2)/2.0
                alphaPairsChanged += 1
                print("iter: %d i:%d, pairs changed %d" % (iter,i,alphaPairsChanged))
        
        if (alphaPairsChanged == 0): iter += 1
        else: iter = 0
        print("iteration number: %d" % iter)
    return b,alphas
```
